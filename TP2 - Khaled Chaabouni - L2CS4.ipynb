```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import cv2 as cv 
import numpy as np
```

```python
img= cv.imread('C:/Users/CHAABOUNI/Desktop/Working.png')
plt.imshow(img)
```

```python
img2 = img[:,:,::-1] 
plt.imshow(img2)
```

```python
print (img2.shape) 
print("Heigh of Image",(img2.shape[0]), "pixels") 
print("Width of Image: ",(img2.shape[1]), "pixels")
```

```python
gray = cv.cvtColor(img2, cv.COLOR_BGR2GRAY) 
print (gray.shape) 
cv.imwrite('WorkingSad.png', gray)
```

```python
equ = cv.equalizeHist(gray)
```

```python
res = np.hstack((gray,equ))  
cv.imwrite('Work.png',res)
```

```python
roihist = cv.calcHist([gray],[0], None, [256], [ 0, 256] ) 
roihist2 = cv.calcHist([equ],[0], None, [256], [ 0, 256] ) 
xs=np.linspace(0,255,256) 
plt.subplot(2,1,1) 
plt.plot(xs,roihist,color='k') 
plt.subplot(2,1,2) 
plt.plot(xs,roihist2,color='r') 
```

```python
color = ('b','g','r') 
plt.subplot(2,1,2) 
for i,col in enumerate(color): 
 histr = cv.calcHist([img2],[i],None,[256],[0,256]) 
 plt.plot(histr,color = col) 
 plt.xlim([0,256]) 
plt.show() 
```
